```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Stock Price Analysis: Nvidia, Palantir and  Tesla

```python
Nvidia_df = pd.read_csv("NVIDIA Stock Price History.csv",na_values = '---')
Nvidia_df
```

```python
PLTR_df = pd.read_csv("Palantir Stock Price History.csv",na_values = '---')
PLTR_df
```

```python
Tesla_df = pd.read_csv("Tesla Stock Price History.csv",na_values = '---')
Tesla_df
```

```python
names = ["Tesla", "Nvidia", "PLTR"]
dfs = [Tesla_df, Nvidia_df, PLTR_df]
```

```python
# Reordering for convinience
for df in dfs:
    df["Date"] = pd.to_datetime(df["Date"])
    df.sort_values("Date", inplace=True)
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
# Adding a returns column
for df in dfs:
    df["Return"] = df["Price"].pct_change()
```

```python
# Core stats
for name, df in zip(names, dfs):
    print(f"\n{name}")
    print("Mean:", df["Return"].mean())
    print("Volatility:", df["Return"].std())
```

```python
# Annualised metric
for name, df in zip(names, dfs):
    mean = df["Return"].mean()
    std = df["Return"].std()
    
    print(f"\n{name}")
    print("Annual Return:", mean * 252)
    print("Annual Volatility:", std * np.sqrt(252))
```

```python
# Key metric Sharpe ratio
for name, df in zip(names, dfs):
    mean = df["Return"].mean()
    std = df["Return"].std()
    
    sharpe = (mean * 252) / (std * np.sqrt(252))
    
    print(f"{name} Sharpe:", sharpe)

# Suggests better return than risks offered by Nvidia stock which probably is a result of the recent AI tech boom and ofcourse tesla is high risk high reward comapred to PLTR.
```

```python
# Normalised Comparison
plt.figure(figsize=(10,6))

for name, df in zip(names, dfs):
    norm = df["Price"] / df["Price"].iloc[0]
    plt.plot(df["Date"], norm, label=name)

plt.legend()
plt.title("Normalised Price Comparison")
plt.xlabel("Date")
plt.ylabel("Normalised Price")
plt.grid(True)
plt.show()
# Illustrates that a wiser choice of investment over extended periods is Nvidia and PLTR.
```

#### Nvidia significantly outperforms both Tesla and Palantir, exhibiting sustained exponential growth, while Tesla shows relatively stagnant performance despite high volatility. Palantir demonstrates strong but less consistent growth, with noticeable drawdowns indicating higher risk.

```python
# Dropping NA values and finding correlation between them
prices = pd.DataFrame({
    "Tesla": Tesla_df.set_index("Date")["Price"],
    "Nvidia": Nvidia_df.set_index("Date")["Price"],
    "PLTR": PLTR_df.set_index("Date")["Price"]
}).dropna()

returns = prices.pct_change()

print(returns.corr())
```

#### The assets exhibit moderate positive correlation (~0.47), indicating that while they are influenced by common market factors, they retain a degree of independent movement, offering limited diversification benefits.

```python
#Drawdowns
drawdown_results = []

for name, df in zip(names, dfs):
    price = df["Price"].dropna()
    
    running_max = price.cummax()
    drawdown = (price - running_max) / running_max
    
    max_dd = drawdown.min()
    
    drawdown_results.append({
        "Asset": name,
        "Max Drawdown": max_dd
    })

drawdown_df = pd.DataFrame(drawdown_results)
display(drawdown_df)
```

#### Maximum drawdown analysis reveals substantial downside risk across all assets, with Palantir experiencing the most severe loss of approximately 85%. Nvidia, while delivering the highest overall returns, demonstrates comparatively lower drawdowns, indicating a more favourable risk-return profile. Tesla exhibits significant volatility with large drawdowns but lacks the consistent upward trend observed in Nvidia

```python
# Return Distributions
plt.figure(figsize=(10,6))

# Use same bins
all_returns = pd.concat([df["Return"].dropna() for df in dfs])
bins = np.linspace(all_returns.min(), all_returns.max(), 60)

for name, df in zip(names, dfs):
    returns = df["Return"].dropna()
    
    plt.hist(returns,
             bins=bins,
             alpha=0.4,
             label=name,
             density=True)

    mean = returns.mean()
    plt.axvline(mean, linestyle='--', linewidth=1, label=f"{name} mean")

plt.title("Distribution of Daily Returns", fontsize=14)
plt.xlabel("Daily Return", fontsize=12)
plt.ylabel("Density", fontsize=12)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.xlim(-0.15, 0.15)

plt.show()
    
```

#### The return distributions are centred around zero but differ in dispersion. Nvidia exhibits a tighter distribution, indicating more stable daily returns, while Tesla and Palantir display wider distributions with heavier tails, reflecting higher volatility and a greater likelihood of extreme price movements. These findings are consistent with the observed drawdowns and highlight the risk-return trade-off across the assets

```python
# Value at risk
var_results = []

for name, df in zip(names, dfs):
    returns = df["Return"].dropna()
    
    var_95 = np.percentile(returns, 5)
    var_99 = np.percentile(returns, 1)
    
    var_results.append({
        "Asset": name,
        "VaR (95%)": var_95,
        "VaR (99%)": var_99
    })

var_df = pd.DataFrame(var_results)
display(var_df)
```

#### Value at Risk (VaR) is used to estimate the potential loss in an asset over a given time period at a specified confidence level. The 95% VaR represents the loss threshold that is only exceeded 5% of the time.

#### The results show that Palantir exhibits the most negative VaR, indicating a higher likelihood of extreme losses. Tesla also demonstrates significant downside risk, while Nvidia shows comparatively less severe losses.

#### These findings are consistent with the earlier analysis of return distributions and maximum drawdowns, reinforcing that assets with higher volatility are more exposed to extreme negative returns

```python
# Sortiono ratiot
sortino_results = []

for name, df in zip(names, dfs):
    returns = df["Return"].dropna()
    
    mean = returns.mean()
    
    # Only downside returns
    downside = returns[returns < 0]
    downside_std = downside.std()
    
    sortino = (mean * 252) / (downside_std * np.sqrt(252))
    
    sortino_results.append({
        "Asset": name,
        "Sortino Ratio": sortino
    })

sortino_df = pd.DataFrame(sortino_results)
display(sortino_df)
```

#### The Sortino ratio refines the Sharpe ratio by considering only downside volatility. This provides a more accurate measure of risk-adjusted performance, particularly for assets with asymmetric return distributions. Nvidia is expected to exhibit a stronger Sortino ratio, reflecting more consistent positive returns relative to downside risk.

## Conclusions

#### Palantir displays the most extreme behaviour, with the widest return distribution, the largest maximum drawdown, and the most negative Value at Risk. These characteristics indicate a highly speculative asset with a greater likelihood of extreme gains and losses.

#### The correlation analysis shows moderate positive relationships between all three assets, suggesting exposure to common market or sector-wide factors. However, the differences in return trajectories and risk metrics demonstrate that individual asset behaviour remains distinct.

#### Overall, the results illustrate the fundamental trade-off between risk and return. While high-growth assets offer the potential for significant gains, they are also associated with substantial downside risk. Among the assets considered, Nvidia provides the most balanced combination of growth and stability, whereas Tesla and Palantir exhibit higher risk with less consistent performance.
